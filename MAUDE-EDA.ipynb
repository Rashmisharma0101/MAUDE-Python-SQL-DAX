from sklearn.base import BaseEstimator, TransformerMixin
class datefunc(BaseEstimator, TransformerMixin):
    def __init__(self, col):
        self.col = col

    def fit(self, X, y = None):
        return self

    def transform(self, X):
        X = X.copy()
        X[self.col] = pd.to_datetime(X[self.col], errors='coerce')
        X[self.col + '_date'] = X[self.col + 'Date'].dt.date
        X[self.col + '_day'] = X[self.col + 'Date'].dt.day
        X[self.col + '_month'] = X[self.col + 'Date'].dt.month
        return X

```python
'''# Create a manufacturer performance summary table with:

Total events

Total severe events

Severe percentage

First event year

Latest event year'''

Manufacturer_Summary = df.groupby('Manufacturer').agg(
    total_events = ('Report Number', 'count'),
    total_severe_events = ('Severe', 'sum'),
    Severe_avg = ('Severe', 'mean'),
    first_event_year = ('Eventdate_year', 'min'),
    last_event_year = ('Eventdate_year', 'max')).reset_index()

Manufacturer_Summary.to_csv('Manufacturer_Summary.csv')
```
